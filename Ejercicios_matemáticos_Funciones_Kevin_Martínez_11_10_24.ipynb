```python
import random

def juego_matematico():
    def generar_ejercicio(operacion):
        num1 = random.randint(1, 10)
        num2 = random.randint(1, 10)

        if operacion == "suma":
            respuesta_correcta = num1 + num2
            simbolo = '+'
        elif operacion == "resta":
            respuesta_correcta = num1 - num2
            simbolo = '-'
        elif operacion == "multiplicacion":
            respuesta_correcta = num1 * num2
            simbolo = '*'
        elif operacion == "division":
            # Aseguramos que la división sea exacta
            num1 = num1 * num2
            respuesta_correcta = num1 // num2
            simbolo = '/'
        else:
            raise ValueError("Operación no soportada.")

        return num1, num2, simbolo, respuesta_correcta

    def realizar_ejercicios(operacion):
        aciertos = 0
        while aciertos < 10:
            num1, num2, simbolo, respuesta_correcta = generar_ejercicio(operacion)
            print(f"¿Cuánto es {num1} {simbolo} {num2}?")
            try:
                respuesta_usuario = int(input("Respuesta: "))
                if respuesta_usuario == respuesta_correcta:
                    print("¡Correcto!")
                    aciertos += 1
                else:
                    print(f"Incorrecto. La respuesta correcta era {respuesta_correcta}.")
            except ValueError:
                print("Por favor, ingresa un número válido.")
        print(f"¡Felicidades! Completaste los 10 ejercicios de {operacion}.")

    def menu():
        print("Bienvenido al juego matemático. Selecciona una opción:")
        print("1. Practicar suma")
        print("2. Practicar resta")
        print("3. Practicar multiplicación")
        print("4. Practicar división")
        print("5. Mezcla de todas las operaciones")

        while True:
            try:
                opcion = int(input("Ingresa una opción (1-5): "))
                if opcion in [1, 2, 3, 4, 5]:
                    return opcion
                else:
                    print("Por favor, elige una opción válida (1-5).")
            except ValueError:
                print("Por favor, ingresa un número válido.")

    def seleccionar_operacion(opcion):
        if opcion == 1:
            realizar_ejercicios("suma")
        elif opcion == 2:
            realizar_ejercicios("resta")
        elif opcion == 3:
            realizar_ejercicios("multiplicacion")
        elif opcion == 4:
            realizar_ejercicios("division")
        elif opcion == 5:
            operaciones = ["suma", "resta", "multiplicacion", "division"]
            for i in range(10):
                operacion = random.choice(operaciones)
                realizar_ejercicios(operacion)

    # Inicio del juego
    opcion = menu()
    seleccionar_operacion(opcion)

# Para ejecutar el programa, solo es necesario llamar a la función:
juego_matematico()
```

```python
juego_matematico()
```
